```python
import numpy as np
import matplotlib.pyplot as plt

# Array
center_red = [(0, 6.0705199999997801e-08), (1, 2.1015700100300739e-08),
     (2, 7.6280656623374823e-09), (3, 5.7348209304555086e-09),
     (4, 3.6812203579604238e-09), (5, 4.1572516753310418e-09)]
center_green = [(0, 6.0705199999997801e-08), (1, 2.1015700100300739e-08),
     (2, 7.6280656623374823e-09), (3, 5.7348209304555086e-09),
     (4, 3.6812203579604238e-09), (5, 4.1572516753310418e-09)]
center_blue = [(0, 6.0705199999997801e-08), (1, 2.1015700100300739e-08),
     (2, 7.6280656623374823e-09), (3, 5.7348209304555086e-09),
     (4, 3.6812203579604238e-09), (5, 4.1572516753310418e-09)]

plt.ion()
data = (center_red, center_green, center_blue)
# colors = ("red", "green", "blue")
colors = ('r--', 'g-', 'bs')
groups = ("Prediction", "Detection", "Update")

# Create plot
fig = plt.figure()
ax = fig.subplots(1, 1)
for cent, color, group in zip(data, colors, groups):
    x_only, y_only = zip(*cent)
    ax.plot(x_only, y_only, color,  alpha=0.8, label=group)

plt.title('Kalman Kilter')
plt.legend(loc=2)
plt.show()
```
